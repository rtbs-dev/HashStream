```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import json
```

```python
times = []
tags = []
with open('./tweet_input/tweets.txt') as f:
    er_no = 0
    
    for tweet in f:
#         print tweet
        dat = json.loads(tweet)
#         print dat.keys()
        
        try:
            times += [dat[u'created_at']]
            
            tags += [[i[u'text'] for i in dat[u'entities'][u'hashtags']]]
        except(KeyError):
#             print 'Missing Timestamp'
            er_no+=1
            pass
    print 'Done!'
    if er_no !=0:
        print 'Dropped {:d} Tweets with missing information!'.format(er_no)
```

```python
df = pd.DataFrame( data = {'time':pd.to_datetime(times), 'hashtags':tags})
# df.ix[4]['HashTags']
sns.distplot([len(i) for i in df.hashtags], kde=False, bins=range(11))
plt.yscale('log')
```

```python
from pandas.tseries.offsets import *
# sec60 = DateOffset(seconds=60)
start = df.time.min()
df['deltaT'] = df.time - start
df['num_tags'] = [len(i) for i in df.hashtags]
df[df.time<df.time[0]+Minute()]
# df.time[0]>df.time[0]+Minute()
```

```python
df[df.time<df.time[0]+Minute()].plot('deltaT','num_tags' )
```

```python
current_tags = []
sorts = df.sort_values('time')
sorts.index=sorts.time
for i in sorts.time:
#     print i, i+Minute()
    temp = sorts[i-Minute():i]
    tot=temp['num_tags'].max()
    current_tags+=[tot]
plt.plot(current_tags)
```
